```python
class calculator():
    def __init__(self,choice1,choice2):
        self.choice1=choice1
        self.choice2=choice2
    def addition(self):
        return  self.choice1 + self.choice2
    def subtraction(self):
        return self.choice1 - self.choice2
    def multiplication(self):
        return self.choice1 * self.choice2
    def division(self):
        return self.choice1 /  self.choice2
    def square(self):
        return self.choice1**2 , self.choice2**2
    def modulus(self):
        return self.choice1 % self.choice2
    def floor_division(self):
        return self.choice1 // self.choice2
    def greater_than(self):
        return self.choice1 > self.choice2
    def less_than(self):
        return self.choice1 < self.choice2
    def equal(self):
        return self.choice1 == self.choice2
    def Max(self):
        return max(self.choice1, self.choice2)
    def Min(self):
        return min(self.choice1, self.choice2)
    def negative(self):
        return self.choice1 < 0 , self.choice2 < 0
    def positive(self):
        return self.choice1 > 0, self.choice2 > 0
    def cube(self):
        return self.choice1**3 , self.choice2**3

choice_1=int(input("Enter first number:  "))
choice_2=int(input("Enter second number: "))

my_calc=calculator(choice_1,choice_2)

choice=1
while choice!=0:
    print()
    print("0. Exit")
    print("1. Addition")
    print("2. Subtraction")
    print("3. Multiplication")
    print("4. Division")
    print("5. Square")
    print("6. Modulus")
    print("7. Floor Division")
    print("8. Greater Than")
    print("9. Less Than")
    print("10. Equal To")
    print("11. Max")
    print("12. Min")
    print("13. Negative")
    print("14. Positive")
    print("15. Cube")
    choice=int(input("Enter your choice... "))
    
    if choice==0:
        break
    elif choice==1:
        print(f"Result: {my_calc.addition()}")
    elif choice==2:
        print(f"Result: {my_calc.subtraction()}")
    elif choice==3:
        print(f"Result: {my_calc.multiplication()}")
    elif choice==4:
        print(f"Result: {my_calc.division()}")
    elif choice==5:
        print(f"Result: {my_calc.square()}")
    elif choice==6:
        print(f"Result: {my_calc.modulus()}")
    elif choice==7:
        print(f"Result: {my_calc.floor_division()}")
    elif choice==8:
        print(f"Result: {my_calc.greater_than()}")
    elif choice==9:
        print(f"Result: {my_calc.less_than()}")
    elif choice==10:
        print(f"Result: {my_calc.equal()}")
    elif choice==11:
        print(f"Result: {my_calc.Max()}")
    elif choice==12:
        print(f"Result: {my_calc.Min()}")
    elif choice==13:
        print(f"Result: {my_calc.negative()}")
    elif choice==14:
        print(f"Result: {my_calc.positive()}")
    elif choice==15:
        print(f"Result: {my_calc.cube()}")
    
    else:
        print("That is an invalid value")
```
